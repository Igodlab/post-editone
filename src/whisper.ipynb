```python
import torch
import numpy

import whisper_timestamped as whisper

import os
```

```python
print(torch.cuda.is_available())
```

```python
help(whisper)
```

```python
# data
data_path = os.path.join('..', 'data')
fname = os.path.join(data_path, 'test00.wav')
```

```python
# load 
audio = whisper.load_audio(fname)
```

```python

model = whisper.load_model("tiny") #, device="cpu")

result = whisper.transcribe(model, audio) #, language="eng")

import json
print(json.dumps(result, indent = 2, ensure_ascii = False))
```

```python
with open(os.path.join(data_path, "test00.json"), "w") as outfile:
    outfile.write(json.dumps(result, indent = 2, ensure_ascii = False))
```
